# **Logistic Model**

```python
# step 1: import library
import pandas as pd
```

```python
# step 2: read dataset
df = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Diabetes.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# step 3: define y (output) and X (input)
```

```python
df.columns
```

```python
y = df['diabetes']
```

```python
X = df[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi',
       'dpf', 'age']]
```

```python
# step 4: train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=2529)
```

```python
# step 5: select logistic regression model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# step 6: train model
model.fit(X_train,y_train)
```

```python
# step 7: prediction
y_pred = model.predict(X_test)
```

```python
# step 8: evaluate your model (accuracy)
from sklearn.metrics import classification_report, confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
import seaborn as sns
sns.regplot(x='glucose',y='diabetes', data = df)
```

```python
import seaborn as sns
sns.regplot(x='glucose',y='diabetes',logistic=True, data = df)
```

# **Decision Tree Model**

```python
# step 1: import library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# step 2: read dataset
df = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Diabetes.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# step 3:define y and X
df.columns
```

```python
y = df['diabetes']
X = df[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi',
       'dpf', 'age']]
```

```python
# step 4: split data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=2529)
```

```python
# step 5: select model
from sklearn.tree import DecisionTreeClassifier, plot_tree
model = DecisionTreeClassifier(max_depth=3)
```

```python
# step 6: train
model.fit(X_train,y_train)
```

```python
# step 7: prediction
y_pred = model.predict(X_test)
```

```python
# step 8: evaluate
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
fig = plt.subplots(figsize=(15,10))
plot_tree(model, feature_names=X.columns,filled=True);
```

```python
# step 1: import library
import pandas as pd
```

```python
# step 2: read dataset
df = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/MultipleDiseasePrediction.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df['prognosis'].value_counts()
```

```python
df['prognosis'].nunique()
```

```python
df.columns
```

```python
# step 3: define y and X
```

```python
y = df['prognosis']
```

```python
X = df.drop(['prognosis'],axis=1)
```

```python
# step 4: split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,stratify=y,random_state=2529)
```

```python
df.shape, X_train.shape, X_test.shape
```

```python
y_train.value_counts()
```

```python
# step 5: select model
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier()
```

```python
# step 6: train model
model.fit(X_train,y_train)
```

```python
# step 7: prediction
y_pred = model.predict(X_test)
```

```python
# step 8: evaluate
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
model = DecisionTreeClassifier(max_depth=10)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.subplots(figsize=(20,15))
plot_tree(model, feature_names=X.columns, filled=True);
```

```python
# accuracy in train sample
model.fit(X_train,y_train)
y_pred = model.predict(X_train)
accuracy_score(y_train,y_pred)
```

```python
# accuracy in test sample
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
accuracy_score(y_test,y_pred)
```
